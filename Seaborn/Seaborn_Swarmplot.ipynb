```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python
tips = sns.load_dataset('tips')
sns.swarmplot(x='size',y='total_bill',data=tips, palette="rocket",hue='sex')
sns.despine(offset=10)
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url

```python
sns.swarmplot(x='day',y='total_bill',data=tips, palette="rocket",hue='sex')
sns.despine(offset=10)
```
