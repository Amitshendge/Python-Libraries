```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

1. Write a program to draw box plot of life expectancy by continent for a data set

```python
data = pd.read_csv('gapminder.csv')
sns.boxplot(x='continent',y='lifeExp',data=data, palette="pastel")
sns.despine(offset=10, trim=True)
```

2. Write a program to draw a box plot of day by tips for a dataset

```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day',y='tip',data=tips, palette="pastel")
sns.despine(offset=10, trim=True)
```
