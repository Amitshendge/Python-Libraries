```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

1. Write a program to draw a violin plot of sex against day for a given dataset

```python
f, ax = plt.subplots(figsize=(6,6))
sns.set_theme(style="whitegrid")
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,linewidth=1,palette='rocket')
sns.despine(left=True,bottom=True)
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset

```python
iris = sns.load_dataset('iris')
iris
```

```python
f, ax = plt.subplots(figsize=(6,6))
sns.set_theme(style="whitegrid")
sns.violinplot(x='species',y='sepal_length',data=iris,split=True,linewidth=1,palette='rocket')
```
