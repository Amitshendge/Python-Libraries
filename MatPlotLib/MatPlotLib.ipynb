4. Write a Python program to draw line charts of the financial data of Alphabet Inc

```python
data = pd.read_csv('appledata.csv',index_col=0)
data_frame = pd.DataFrame(data)
data_frame
```

```python
plt.plot(data_frame)
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line

```python
plt.plot(data_frame)
plt.legend(data_frame.columns)
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
```

6. Write a Python program to plot two or more lines with legends, different widths and colors.


```python
plt.plot(data_frame['Open'],lw=3,color='b')
plt.plot(data_frame['High'],color='r',lw=6)

plt.legend(data_frame.columns)
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
```

7. Write a Python program to plot two or more lines with different styles

```python
plt.plot(data_frame['Open'],color='b',lineStyle='dotted')
plt.plot(data_frame['Low'],color='r',lineStyle='dashed')

plt.legend(['Open','Low'])
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
```

8. Write a Python program to plot two or more lines and set the line markers

```python
plt.plot(data_frame['Open'],color='b',marker='o')
plt.plot(data_frame['Close'],color='r',marker='x')

plt.legend(['Open','Close'])
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
```

9. Write a Python program to display the current axis limits values and set new axis values

```python
plt.plot(data_frame['Open'],color='b',marker='o')
plt.plot(data_frame['Close'],color='r',marker='x')

plt.legend(['Open','Close'])
plt.xlabel('Date')
plt.ylabel('Price is $')
plt.title('Alphabet Financials')
plt.axis([-1, 5, 770, 785])
```

12. Write a Python program to create multiple types of charts

```python
fig,axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)

axs[0].bar(data_frame.index,data_frame['Close'])
axs[1].scatter(data_frame.index,data_frame['Close'])
axs[2].plot(data_frame.index,data_frame['Close'])
```

13. Write a Python program to display the grid and draw line charts of the closing value of

```python
plt.plot(data_frame['Close'],color='r')
plt.grid(color='b', linestyle='-', linewidth=0.5)
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc

```python
plt.plot(data_frame['Close'],color='b')
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.tick_params(which='both',top='off',left='off',right='off',bottom='off')
```

15. Write a Python program to create multiple plots

```python
x = np.array([1,2,3,4,5,6])
fig,axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)

axs[0].plot(x,x*x)
axs[1].plot(x,x*x*x)
axs[2].plot(x,x*x*x*x)
```
